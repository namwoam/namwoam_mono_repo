## 1. Matrix Multiplication and Solve a Linear System

### Introduction
1. Use ```np.array``` to construct $A$ and $B$.
2. Use ```np.linalg.solve``` to solve $AX = B$.
3. Use ```np.linalg.inv``` to get $A^{-1}$.
4. Output the matrix / vector.

````python
import numpy as np
from matplotlib import pyplot as plt
# (a)
dataA = [[-5, 4, 0, 1],
         [-30, 27, 2, 7],
         [5, 2, 0, 2],
         [10, 1, -2, 1],
         ]
dataB = [-17, -102, -7, 6]
A = np.array(dataA)
B = np.array(dataB)
X = np.linalg.solve(A, B)
# (b)
print(str(X))
inverseA = np.linalg.inv(A)
# (c)
print(inverseA)
# (d)
print(np.matmul(inverseA, B))
````

### Conclusion
1. We get the solution of $AX = B$ , $X = \begin{bmatrix} 2.2 \\ 1 \\ 3.5 \\ -10 \end{bmatrix}$
2. We get the inverse matrix $A^{-1} = 
\begin{bmatrix} .
-0.66 & 0.10 & -0.066 & 0.10  \\ 
-1.33 & 0.25 & -0.33 & 0.25 \\ 
-2.50 & 0.375 & 0.00 & -0.125 \\ 
-3.00 & -0.50 & 1.00 & -0.50 \\ 
\end{bmatrix}$.
3. Since $AX=B$ , therefore $A^{-1}B = A^{-1}AX = X$.

## 2.  Image Processing and Filters

### Introduction
1. Use ```cv2.imread``` to load images.
2. Use ```cv2.add``` and ```cv2.subtract``` to perform arithmetic operations on images.
3. Use ```cv2.getRotationMatrix2D``` to get rotation matrix and use ```cv2.warpAffine``` to apply image rotation.
4. Use ```cv2.Canny``` to perform Canny Edge Detection.

````python
import cv2
import matplotlib.pyplot as plt
raw = cv2.imread("person.jpg")
rawShape = (raw.shape[1], raw.shape[0])
pika = cv2.imread("pika.jpg")
A = cv2.add(raw, cv2.resize(pika, rawShape))
icon = cv2.imread("thumb.png")
B = cv2.subtract(raw, cv2.resize(icon, rawShape))
rotationMatrix = cv2.getRotationMatrix2D( (rawShape[0]/2, rawShape[1]/2 ) , -60, 2)
C = cv2.warpAffine(raw  , rotationMatrix , rawShape)
D = cv2.Canny(raw , 100 , 200)



imageDisplay=[A, B , C , D]
for i in range(len(imageDisplay)):
    plt.axis("off")
    plt.imshow(cv2.cvtColor(imageDisplay[i], cv2.COLOR_BGR2RGB))
    if (i != len(imageDisplay) - 1):
        plt.figure()
````
